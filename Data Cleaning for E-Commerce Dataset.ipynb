```python
import pandas as pd

df = pd.read_csv("data.csv", encoding="latin1", dtype={
    "StockCode": str,
    "InvoiceNo": str,
    "CustomerID": str
})

# function to replace non-numeric with mode
def replace_with_mode(column):
    # find numeric values only
    numeric_values = df[df[column].str.contains(r'^\d+$', na=False)][column]
    
    # find mode
    mode_value = numeric_values.mode()[0]
    
    print(f"Mode for {column}:", mode_value)
    
    # replace non-numeric values with mode
    df.loc[~df[column].str.contains(r'^\d+$', na=False), column] = mode_value

# apply to all columns
replace_with_mode("StockCode")
replace_with_mode("InvoiceNo")
replace_with_mode("CustomerID")

# verify
print("\nCheck datatypes:")
print(df.dtypes)

print("\nSample data:")
print(df.head())

# save cleaned file
df.to_csv("clean_data.csv", index=False)
```
